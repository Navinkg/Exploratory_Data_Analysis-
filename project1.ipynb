3.What is the average birth weight and length of babies ? 

```python
avg_weight = df['birth_weight_kg'].mean()
avg_length = df['birth_length_cm'].mean()
print('average weight :',avg_weight)
print('average lenth:',avg_length)
```

```python
def convert_weight(val):
    if 1 < val <= 2:
        return '1_to_2_kg'
    elif 2 < val <= 3:
        return '2_to_3_kg'
    elif 3 < val <= 4:
        return '3_to_4_kg'
    elif 4 < val <= 5:
        return '4_to_5_kg'
df['birth_wt_groups'] = df['birth_weight_kg'].apply(convert_weight)
df.groupby('gender')['birth_wt_groups'].value_counts()
```

```python
def convert_length(val):
    if val < 45:
        return 'Below_45_cm'
    elif 45 <= val < 50:
        return '45_to_50_cm'
    elif 50 <= val < 55:
        return '50_to_55_cm'
    else:
        return 'Above_55_cm'
df['length_groups'] = df['birth_length_cm'].apply(convert_length)
df.groupby('gender')['length_groups'].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
sns.countplot(
    data=df, 
    x='birth_wt_groups', 
    hue='gender',
    order=['1_to_2_kg','2_to_3_kg','3_to_4_kg','4_to_5_kg']
)
plt.title("Birth Weight Groups by Gender")
plt.xlabel("Weight Group")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(
    data=df,
    x='length_groups',
    hue='gender',
    order=['Below_45_cm','45_to_50_cm','50_to_55_cm','Above_55_cm']
)
plt.title("Birth Length Groups by Gender")
plt.xlabel("Length Group (cm)")
plt.ylabel("Count")
plt.show()
```

The analysis shows that the average birth weight and average birth length of newborns fall within the expected healthy range. These values help us understand the general growth patterns in the dataset and provide a baseline for identifying babies who are underweight or shorter than typical newborn measurements.
